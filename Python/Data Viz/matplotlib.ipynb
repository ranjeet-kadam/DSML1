```python
import matplotlib.pyplot as plt
```

```python
x= [1,2,3,4,5]
y=[1,4,9,16,25]

# Create a line plot
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic line plot")
plt.show()
```

```python
## Create a customize plot

x= [1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y,color='blue',linestyle='--',marker='o')
plt.grid(True)
```

```python
### Multiple plot 

x= [1,2,3,4,5]
y=[1,4,9,16,25]
z =[1,8,27,64,125]

plt.figure(figsize=(8,7))

plt.subplot(1,2,1)
plt.plot(x,y,color='purple')
plt.title('Plot A')

plt.subplot(1,2,2)
plt.plot(x,z,color='blue')
plt.title('Plot B')
```

```python
### Bar Plot

categories = ['T-shirts','Pants','Shirts','Sports Lower','Gym Shoes']
Sales = [214,342,123,675,210]

plt.bar(categories,Sales,color='purple')
plt.xlabel('Categories')
plt.ylabel('Sales')
```

```python
### Histogram

data = [1,2,2,3,3,3,4,4,5,5,5]

plt.hist(data,bins=5,color='violet',edgecolor='black')

```

```python
### Scatter Plot

x=[1,2,3,4,5]
y=[4,5,6,7,8]

plt.scatter(x,y,color='Red',marker='*')
```

```python
### Pie Chart 

labels=['Sales','Marketing','R & D','Product Developement']
contribution = [20,15,35,30]
colors = ['gold','lightcoral','yellowgreen','lightskyblue']
expload=(0.2,0,0,0)

plt.pie(contribution,labels=labels,colors=colors,autopct="%1.1f%%")
```
