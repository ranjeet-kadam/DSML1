```python
import pandas as pd
```

```python
df = pd.read_csv("Merged_Sales_Data.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
### Handlling missing values

df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df_fill = df.fillna(0)

df_fill.isnull().sum()
```

```python
### Missing Value with Mean
df['Price Each']=pd.to_numeric(df['Price Each'],errors='coerce')
df['Price']=df['Price Each'].fillna(df['Price Each'].mean())
df
```

```python
df = df.rename(columns={'Purchase Address':'Address'})
df.head()
```

```python
### Change Datatypes

df['Price1']=df['Price'].astype(int)
df
```

```python
### Group by and aggregate function

grouped_product = df.groupby('Product')['Price'].sum()
print(grouped_product)
```

```python
grouped_product = df.groupby('Product')['Price'].mean()
print(grouped_product)
```

```python
### Apply multiple aggregate function

grouped_agg = df.groupby('Product')['Price'].agg(['mean','sum','count'])
grouped_agg
```

```python
### Merge and Join operation

df1 = pd.DataFrame({'Key':['A','B','C'], 'Value1':[1,2,3]})
df2 = pd.DataFrame({'Key':['A','B','D'], 'Value1':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
### Merge data by using inner join

pd.merge(df1,df2,on='Key',how='inner',suffixes=('_','_'))
```

```python
### Merge data by using outer join

pd.merge(df1,df2,on='Key',how='outer',suffixes=('_','_'))
```

```python
### Merge data by using left join

pd.merge(df1,df2,on='Key',how='left',suffixes=('_','_'))
```

```python
### Merge data by using right join

pd.merge(df1,df2,on='Key',how='right',suffixes=('_','_'))
```

```python
### Join operation
df1 = pd.DataFrame({'Name': ['Alice', 'Bob']}, index=[1, 2])
df2 = pd.DataFrame({'Score': [85, 90]}, index=[1, 2])
df1.join(df2)
```
