```python
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({
    'color':['red','blue','yellow,','green','beige','aqua blue','red','blue','yellow,','green','beige','aqua blue']
})

df
```

```python
### Create an instance for OHE

encoder = OneHotEncoder()
```

```python
### Perform fit and transform

encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
df_encoded=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
df_encoded
```

```python
pd.concat([df,df_encoded],axis=1)
```
